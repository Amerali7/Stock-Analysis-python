```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2017,3,9)
```

```python
tesla = web.DataReader("TSLA",'yahoo',start,end)
ford = web.DataReader("F",'yahoo',start,end)
gm = web.DataReader("GM",'yahoo',start,end)
```

```python
tesla.to_csv('Tesla_stock.csv')
ford.to_csv('Ford_stock.csv')
```

```python
tesla.head()
```

```python
ford.head()
```

```python
gm.head()
```

```python
tesla['Open'].plot(label='Tesla', figsize=(15,7))
gm['Open'].plot(label='gm')
ford['Open'].plot(label='ford')
plt.ylabel('stock price')
plt.title('stock analysis')
plt.legend()
                   
```

```python
tesla['Volume'].plot(label='Tesla',figsize=(17,5))
gm['Volume'].plot(label='GM')
ford['Volume'].plot(label='Ford')
plt.legend()
```

```python
start = datetime.datetime(2022,1,1)
end = datetime.datetime(2022,3,9)
```
